```
X_train1 = np.load('Grad1.npy')
X_train2 = np.load('Grad2.npy')
X_train3 = np.load('Grad3.npy')
X_train4 = np.load('Grad4.npy')
X_train5 = np.load('Grad5.npy')
X_train1 = np.concatenate((X_train1, X_train2), axis = 0)
X_traina = np.concatenate((X_train1, X_train3, X_train4, X_train5), axis = 0)
X_train1_train = X_train1[:265]
X_train1_val = X_train1[265:]
X_train3_train = X_train3[:145]
X_train3_val = X_train3[145:]
X_train4_train = X_train4[:54]
X_train4_val = X_train4[54:]
X_train5_train = X_train5[:47]
X_train5_val = X_train5[47:]

print(X_train1.shape)

print(X_train3.shape)
print(X_train4.shape)
print(X_train5.shape)
print(X_traina.shape)
```

```
def y_out(y, num_classes = 4):
  y1 = np.zeros((y.shape[0],num_classes))
  for i in range(len(y)):
    y1[i,y[i]-1] = 1
  
  return y1
y_train1 = np.concatenate((y_train1, y_train2), axis = 0)
y_train1 = y_out(y_train1)
y_train3 = y_out(y_train3)
y_train4 = y_out(y_train4)
y_train5 = y_out(y_train5)
X_train = np.concatenate((X_train1_train, X_train3_train, X_train4_train, X_train5_train), axis = 0)
X_val = np.concatenate((X_train1_val, X_train3_val, X_train4_val, X_train5_val), axis = 0)
y_train1_train = y_train1[:265]
y_train1_val = y_train1[265:]
y_train3_train = y_train3[:145]
y_train3_val = y_train3[145:]
y_train4_train = y_train4[:54]
y_train4_val = y_train4[54:]
y_train5_train = y_train5[:47]
y_train5_val = y_train5[47:]
y_train = np.concatenate((y_train1_train, y_train3_train, y_train4_train, y_train5_train), axis = 0)
y_val = np.concatenate((y_train1_val, y_train3_val, y_train4_val, y_train5_val), axis = 0)
print(X_train.shape)
print(y_train.shape)
print(X_val.shape)
print(y_val.shape)
```

```
import keras
import numpy as np
import matplotlib.pyplot as plt
from keras import layers
from keras import optimizers
from keras.layers import Dense, Flatten, Activation, Dropout
from keras import applications
from keras.applications import VGG16, InceptionV3, ResNet50

image_size = 224

vgg_conv = ResNet50(weights = 'imagenet', include_top = False, input_shape=(image_size, image_size, 3))

vgg_conv.summary()
```

```
weights_path = 'resnet50_weights_tf_dim_ordering_tf_kernels_notop.h5'
```

```
from keras.models import Sequential, Model, load_model

vgg_conv.load_weights(weights_path)

x = Flatten()(vgg_conv.get_layer('activation_49').output)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(4, activation='softmax')(x)

model = Model(inputs = vgg_conv.input, outputs = x)
model.compile(loss='categorical_crossentropy', optimizer = optimizers.Adam(lr=0.00004, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False), metrics = ['accuracy'])

model.summary()
```

```
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ModelCheckpoint

batch_size = 20
epochs = 160

train_datagen = ImageDataGenerator(rotation_range = 10, width_shift_range = 0.1, height_shift_range = 0.1, horizontal_flip=True)

history = model.fit_generator(train_datagen.flow(X_train, y_train, batch_size = batch_size),
                             steps_per_epoch = X_train.shape[0] // batch_size,
                             epochs = epochs,
                             validation_data = (X_val, y_val),
                             callbacks=[ModelCheckpoint('ResNet50-transferlearning.model', monitor='val_acc', save_best_only=True)])
```
